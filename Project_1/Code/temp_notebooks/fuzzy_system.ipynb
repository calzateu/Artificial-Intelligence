```python
import numpy as np
import skfuzzy as fuzz
import skfuzzy.membership as mf
import matplotlib.pyplot as plt
```

# Defining input ranges

First I define variables to control the ranges

```python
max_amount = 800000
min_amount = 1000
```

Generate universe variables

```python
x_transaction_amount = np.arange(min_amount, max_amount, 1)
x_hour = np.arange(0, 23, 0.5)
x_transactions_per_day = np.arange(0, 100, 1)
x_transactions_per_week = np.arange(0, 100, 1)
y_risk = np.arange(0, 1, 0.01)
```

# Defining membership functions

```python
transaction_amount_lo = fuzz.gaussmf(x_transaction_amount, 0, 140000)
transaction_amount_md = fuzz.gaussmf(x_transaction_amount, 400000, 140000)
transaction_amount_hi = fuzz.gaussmf(x_transaction_amount, 800000, 140000)

hour_early_morning = fuzz.gaussmf(x_hour, 0, 5)
hour_day = fuzz.gaussmf(x_hour, 12, 5)
hour_night = fuzz.gaussmf(x_hour, 23, 5)

transactions_per_day_lo = fuzz.gaussmf(x_transactions_per_day, 0, 50)
transactions_per_day_hi = fuzz.gaussmf(x_transactions_per_day, 100, 50)

transactions_per_week_lo = fuzz.gaussmf(x_transactions_per_week, 0, 50)
transactions_per_week_hi = fuzz.gaussmf(x_transactions_per_week, 100, 50)

risk_lo = fuzz.gaussmf(y_risk, 0, 0.2)
risk_md = fuzz.gaussmf(y_risk, 0.5, 0.25)
risk_hi = fuzz.gaussmf(y_risk, 1, 0.2)
```

### Plot the functions

```python
# Visualize these universes and membership functions
fig, (ax0, ax1, ax2, ax3, ax4) = plt.subplots(nrows=5, figsize=(8, 17))

ax0.plot(x_transaction_amount, transaction_amount_lo, 'b', linewidth=1.5, label='Low')
ax0.plot(x_transaction_amount, transaction_amount_md, 'g', linewidth=1.5, label='Medium')
ax0.plot(x_transaction_amount, transaction_amount_hi, 'r', linewidth=1.5, label='High')
ax0.set_title('Transaction Amount')
ax0.legend()

ax1.plot(x_hour, hour_early_morning, 'b', linewidth=1.5, label='Early Morning')
ax1.plot(x_hour, hour_day, 'g', linewidth=1.5, label='Day')
ax1.plot(x_hour, hour_night, 'r', linewidth=1.5, label='Night')
ax1.set_title('Hour of Day')
ax1.legend()

ax2.plot(x_transactions_per_day, transactions_per_day_lo, 'b', linewidth=1.5, label='Low')
ax2.plot(x_transactions_per_day, transactions_per_day_hi, 'r', linewidth=1.5, label='High')
ax2.set_title('Transactions per Day')
ax2.legend()

ax3.plot(x_transactions_per_week, transactions_per_week_lo, 'b', linewidth=1.5, label='Low')
ax3.plot(x_transactions_per_week, transactions_per_week_hi, 'r', linewidth=1.5, label='High')
ax3.set_title('Transactions per Week')
ax3.legend()

ax4.plot(y_risk, risk_lo, 'b', linewidth=1.5, label='Low')
ax4.plot(y_risk, risk_md, 'g', linewidth=1.5, label='Medium')
ax4.plot(y_risk, risk_hi, 'r', linewidth=1.5, label='High')
ax4.set_title('Risk')
ax4.legend()
```
